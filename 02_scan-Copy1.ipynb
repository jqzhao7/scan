# 一. 读取文件
- 注意jupyter无法使用cv2显示图片, 因此使用matplotlib, 注意加入%
- `cv2.cvtColor(img,cv2.COLOR_BGR2RGB)` 因为cv2读取的图像是BGR格式的图像,因此需要转化为RGB才能使用matplotlib显示

````python
import cv2
import numpy as np
np.set_printoptions(threshold=np.inf)
from matplotlib import pyplot as plt # 用来解决不能显示的问题
%matplotlib inline
img = cv2.imread(r"imgs/02.png") # 
img_show = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)  
plt.imshow(img_show), plt.title("blue")
print(img.shape)
````

# 二. 下面的操作是对图像进行边缘提取

## 2.1 图像二值化
- 原本图像是彩色三通道, 经过二值化后图像变成黑白0-255, 以行列表示的二维矩阵

````python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY);   # 二值化
img_show = cv2.cvtColor(img_gray,cv2.COLOR_BGR2RGB)  # 转化为plt可以显示的格式
plt.imshow(img_show), plt.title("img_gray")
print(img_gray[1])
````

## 2.2 高斯变换 
https://blog.csdn.net/godadream/article/details/81568844
对图像进行平滑, 保留图像的总体分布特征

````python
img_guassian = cv2.GaussianBlur(img_gray, (5,5), 0)
img_show = cv2.cvtColor(img_guassian,cv2.COLOR_BGR2RGB)  # 转化为plt可以显示的格式
plt.imshow(img_show), plt.title("img_guassian")
````

## 2.3 边缘膨胀
- 形态学边界提取
- 就是先膨胀, 然后在
https://www.cnblogs.com/qianxia/p/11105883.html 
```
膨胀算法：变胖
    用kernel，扫描图像的每一个像素；用kernel与其覆盖的二值图像做 “与” 操作；
    如果都为0，结果图像的该像素为0；否则为1.
    结果：使二值图像扩大一圈
```

````python
# 这里不确定cv2.threshold阈值处理是对灰度图还是对高斯平滑之后进行哪个效果更好
ret, img_binary = cv2.threshold(img_gray, 0, 255, cv2.THRESH_OTSU | cv2.THRESH_BINARY) #二值化后的图转化为和白图

# ret, img_binary = cv2.threshold(img_guassian, 0, 255, 
#                                 cv2.THRESH_OTSU | cv2.THRESH_BINARY) #二值化后的图转化为和白图

plt.subplot(2,2,1)
img_show = cv2.cvtColor(img_binary,cv2.COLOR_BGR2RGB)  # 转化为plt可以显示的格式
plt.imshow(img_show), plt.title("img_binary")
# cv2.imshow("binary", binary)
# 第一个参数 矩形：MORPH_RECT;交叉形：MORPH_CROSS;椭圆形：MORPH_ELLIPSE;
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5)) 
img_dst = cv2.dilate(img_binary, kernel) # 正式膨胀
plt.subplot(2,2,2)
img_show = cv2.cvtColor(img_dst,cv2.COLOR_BGR2RGB)  # 转化为plt可以显示的格式
plt.imshow(img_show), plt.title("img_dst")
````

## 2.5 边缘提取

````python
# https://blog.csdn.net/wsp_1138886114/article/details/82935839
img_canny = cv2.Canny(img_dst, 30, 120)  # 边缘提取常用函数
img_show = cv2.cvtColor(img_canny,cv2.COLOR_BGR2RGB)  # 转化为plt可以显示的格式
plt.imshow(img_show), plt.title("img_canny")
````

## 2.6 找出图像中的最大轮廓并填充
https://blog.csdn.net/qq_33757398/article/details/96174768
- 得到轮廓集合->计算每个轮廓的面积->找到最大面积编号->得到轮廓直线->得到直线交点->透视变换得到新图

````python
img_canny = cv2.Canny(img_dst, 30, 120)  # 边缘提取常用函数
contours, _ = cv2.findContours(img_canny, 
                               cv2.RETR_EXTERNAL, # 只检测外围轮廓
                               cv2.CHAIN_APPROX_SIMPLE) # 保存轮廓上的所有点
# 轮廓的点集(contours), 各层轮廓的索引(hierarchy)
area = [cv2.contourArea(contour) for contour in contours] # 得到每一个轮廓的面积
max_idx = np.argmax(np.array(area))
for i in range(len(img_canny)): # 先清空图
    for j in range(len(img_canny[i])):
        img_canny[i][j] = 0
for i in contours[max_idx]: # 将轮廓点插入到图中
    img_canny[i[0][1]][i[0][0]]  = 255
img_show = cv2.cvtColor(img_canny,cv2.COLOR_BGR2RGB)  # 转化为plt可以显示的格式
plt.imshow(img_show), plt.title("img_canny")
# print(contours[max_idx][:10])
print(area)
print("最大面积的廓落编号",max_idx,area[max_idx])
````

## 绘制的轮廓内填充颜色用于显示

````python
# img_draw = cv2.fillConvexPoly(img_canny, contours[max_idx], 100) # 填充凸聚廓
# 虽然绘制出来的上面有一条横线, 那是因为img_canny图像中包含的, 并不是轮廓
img_draw = cv2.drawContours(img_canny,  # 要绘制轮廓的图像
                            contours, # 所有轮廓
                            max_idx, # 轮廓编号
                            256,  # 轮廓颜色
                            cv2.FILLED)# 填充最大的轮
img_show = cv2.cvtColor(img_draw,cv2.COLOR_BGR2RGB)  # 转化为plt可以显示的格式
plt.subplot(2,2,1)
plt.imshow(img_show), plt.title("img_draw")

img_ = cv2.Canny(img_canny, 30, 120) # 再一次找轮廓就舒服了
img_show = cv2.cvtColor(img_,cv2.COLOR_BGR2RGB)
plt.subplot(2,2,2)
plt.imshow(img_show), plt.title("img_canny")
````

# 霍夫变换求四条直线
 https://blog.csdn.net/qq_40962368/article/details/81950178
 https://blog.csdn.net/ftimes/article/details/106816736
- lines=cv2.HoughLines（image,rho,theta,threshold） # 这里需要整理

````python
img_ = cv2.Canny(img_canny, 30, 120) # 再一次找轮廓就舒服了
img_houghlines = cv2.HoughLinesP(img_, 
                                 2, # 像素点精度
                                 (np.pi / 180),
                                 118, 
                                 minLineLength=100, # 线长
                                 maxLineGap=5)   # 共线两点之间最短距离
print(img_houghlines)
img_houghlines = list(img_houghlines[:, 0, :])  # 将数据转换到二维
# print("转化二维",img_houghlines)
for x1,y1,x2,y2 in img_houghlines: # 在原图上绘制 
    cv2.line(img,(x1,y1),(x2,y2),(0,255,0),2)
img_show = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_show), plt.title("img")
#函数将通过步长为1的半径和步长为π/180的角来搜索所有可能的直线
from src import get_line_cross_point
cross_point = []
for i, (poin1) in enumerate(img_houghlines):
    if i==3:
        break
    for j, (poin2) in enumerate(img_houghlines[i+1:]):
        _ = get_line_cross_point(poin1, poin2)
        if _:cross_point.append(_)
print("得到的交点的结果是",cross_point)
if len(cross_point) != 4:
    print("目标定位失败, 请修改参数后重新尝试")
````

````python
# 这里处理cross_point
print(cross_point)
X_max = 300
Y_max = 400
x = []
y = []
points2 = []
for point in cross_point:
    x.append(point[0])
    y.append(point[1])
x_mean = int(np.mean(x))
y_mean = int(np.mean(y))

for i,point in enumerate(cross_point):
    if point[0] > x_mean:
        points2.append([X_max,0])
    else:
        points2.append([0,0])
    if point[1] > y_mean:
        points2[i][1] = Y_max
print("points2",points2)
````

````python
# rows,cols,_ = img.shape
points1 = np.float32(cross_point) 

# 这四个点就是轮廓求出来的[[40, 383], [278, 358], [40, 65], [253, 79]]
points2 = np.float32(points2)
matrix = cv2.getPerspectiveTransform(points1,points2)
# 将四个点组成的平面转换成另四个点组成的一个平面
# output = cv2.warpPerspective(img, matrix, (cols, rows))
output = cv2.warpPerspective(img, matrix, (300, 400))
# 通过warpPerspective函数来进行变换
img_show = cv2.cvtColor(output,cv2.COLOR_BGR2RGB)  
plt.imshow(img_show), plt.title("output")
````
