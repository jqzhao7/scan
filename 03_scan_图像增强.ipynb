```python
import cv2
import numpy as np
np.set_printoptions(threshold=np.inf)
from matplotlib import pyplot as plt # 用来解决不能显示的问题
%matplotlib inline
img = cv2.imread(r"swap/output.jpg") # 
img_show = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)  
plt.imshow(img_show), plt.title("blue")
print(img.shape)
```

```python
a=2
#线性变换  定义float类型
O = float(a)*img
#数据截取  如果大于255 取 255
O[0>200] = 255
#数据类型的转换
O = np.round(O)
O = O.astype(np.uint8)
img_show = cv2.cvtColor(O,cv2.COLOR_BGR2RGB)  
plt.imshow(img_show), plt.title("blue")
print(img.shape)
cv2.imwrite("swap/O.jpg",O)
```
